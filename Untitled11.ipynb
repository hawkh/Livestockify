```python
import torch
import torchvision
import onnx
import os

# Load your trained YOLOv11 model
model = torch.load("/content/runs/detect/train/weights/best.pt", map_location=torch.device('cpu'),weights_only=True)['model'].float()
model.eval()

# Dummy input for export
dummy_input = torch.randn(1, 3, 640, 640)  # adjust if you used different input size

# Export to ONNX
torch.onnx.export(
    model,
    dummy_input,
    "best.onnx",
    opset_version=12,
    input_names=['images'],
    output_names=['output'],
    dynamic_axes={'images': {0: 'batch'}, 'output': {0: 'batch'}}
)

print("✅ Exported to best.onnx")
```
